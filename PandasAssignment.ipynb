# Assignment 3

    Import the MELBOURNE_HOUSE_PRICES_LESS.csv dataset
    Extract all the values where price column is not null from the dataset to a new variable
    Extract all the values where price column is null from the dataset to a new variable
    Clean the dataset so it deosn't have any null value again using any method
    Compute the sum, mean of the price column on the dataset
    Get the house with the lowest price and with the highest price



```python
# Import the MELBOURNE_HOUSE_PRICES_LESS.csv dataset
melHousePrices = pd.read_csv('C:\PythonClass\Data science\Datasets\MELBOURNE_HOUSE_PRICES_LESS.csv')
melHousePrices[0:21]
```

```python
# Extract all the values where price column is not null from the dataset to a new variable
# price = melHousePrices['Price'][melHousePrices['Price'].notnull()]
prices = melHousePrices['Price'].notnull()
prices = melHousePrices['Price'][prices]
prices
```

```python
# Extract all the values where price column is null from the dataset to a new variable
price = melHousePrices['Price'].isnull()
price = melHousePrices['Price'][price]
price
```

```python
# Clean the dataset so it deosn't have any null value again using any method
melHousePrices.isnull().any()
```

```python
# To clean the data, for the prices missing, the average of each type (h, t, u) is calculated first
# Then the average is multiplied by the number of rooms in that place to get the price
h_prices= melHousePrices.loc[melHousePrices["Type"]=='h', "Price"]
h_price_available = melHousePrices.loc[melHousePrices["Type"]=='h', "Price"].notnull()
h_price_not_available = melHousePrices.loc[melHousePrices["Type"]=='h', "Price"].isnull()
h_rooms = melHousePrices.loc[melHousePrices["Type"]=='h', "Rooms"]
h_prices_sum = h_prices.sum()
h_rooms_sum = h_rooms[h_price_available].sum()
mean_price_h_per_room = h_prices_sum/h_rooms_sum
mean_price_h_per_room

t_prices= melHousePrices.loc[melHousePrices["Type"]=='t', "Price"]
t_price_available = melHousePrices.loc[melHousePrices["Type"]=='t', "Price"].notnull()
t_price_not_available = melHousePrices.loc[melHousePrices["Type"]=='t', "Price"].isnull()
t_rooms = melHousePrices.loc[melHousePrices["Type"]=='t', "Rooms"]
t_prices_sum = t_prices.sum()
t_rooms_sum = t_rooms[t_price_available].sum()
mean_price_t_per_room = t_prices_sum/t_rooms_sum
mean_price_t_per_room

u_prices= melHousePrices.loc[melHousePrices["Type"]=='u', "Price"]
u_price_available = melHousePrices.loc[melHousePrices["Type"]=='u', "Price"].notnull()
u_price_not_available = melHousePrices.loc[melHousePrices["Type"]=='u', "Price"].isnull()
u_rooms = melHousePrices.loc[melHousePrices["Type"]=='u', "Rooms"]
u_prices_sum = u_prices.sum()
u_rooms_sum = u_rooms[u_price_available].sum()
mean_price_u_per_room = u_prices_sum/u_rooms_sum
mean_price_u_per_room

melHousePrices.loc[melHousePrices["Type"]=='u', "Price"] = melHousePrices.loc[melHousePrices["Type"]=='u', "Price"].fillna(mean_price_u_per_room*u_rooms[u_price_not_available])
melHousePrices.loc[melHousePrices["Type"]=='t', "Price"] = melHousePrices.loc[melHousePrices["Type"]=='t', "Price"].fillna(mean_price_t_per_room*t_rooms[t_price_not_available])
melHousePrices.loc[melHousePrices["Type"]=='h', "Price"] = melHousePrices.loc[melHousePrices["Type"]=='h', "Price"].fillna(mean_price_h_per_room*h_rooms[h_price_not_available])
```

```python
melHousePrices.isnull().any()
```

```python
# Compute the sum, mean of the price column on the dataset
print('The sum of the price is: ', melHousePrices['Price'].sum())
print('The mean of the price is: ', melHousePrices['Price'].mean())
```

```python
# Get the house with the lowest price and with the highest price
print('The house with the highest price is: \n', melHousePrices.iloc[melHousePrices['Price'].argmax()], '\n')
print('The house with the lowest price is: \n', melHousePrices.iloc[melHousePrices['Price'].argmin()])
```
